```python
## LINEAR INTERPOLATION
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
```

```python
y = np.array([2,3,6,8,10])
```

```python
x
```

```python
y
```

```python
## INTERPOLATE THE DATA USING THE LINEAR INTERPOLATION
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.scatter(x,y)
```

```python
## INTERPOLATE THE DATA USING THE LINEAR INTERPOLATION
x_new = np.linspace(1,5,10)  ## create new values
y_intrep = np.interp(x_new,x,y)  ## interpolate y values
```

```python
y_intrep
```

```python
plt.scatter(x_new,y_intrep)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
x
```

```python
y
```

```python
from scipy.interpolate import interp1d
```

```python
## create a cubic interpolation function
f = interp1d(x,y,kind='cubic')
```

```python
f
```

```python
## interpolate the data 
x_new = np.linspace(1,5,10)
y_interp =f(x_new)
print(y_intrep)
```

```python
plt.scatter(x_new , y_intrep)
```

```python
## polynomial interpolatin
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
# interpolate the data using polynomial interpolation
p = np.polyfit(x,y,100)  # fit a 2nd degree polynomial to the data  set degree by your choice 
```

```python
p
```

```python
x_new = np.linspace(1,5,100) # crearte new x values 
y_intrep = np.polyval(p,x_new) ## interpolate y values
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_intrep)
```
